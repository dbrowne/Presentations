```julia
# julia set
# (the familiar mandelbrot set is obtained by setting c==z initially)
function julia(z, c; maxiter=200)
    for n = 1:maxiter
        if abs2(z) > 4
            return n-1
        end
        z = z*z + c
    end
    return maxiter
end
# Note:  The explicit function name (actually optional) with a mandatory end to terminate the function scope and if block scope
```

```julia
# The @time macro will give us some basic performance info
# varying the second argument to julia() tiny amounts results in a stunning variety of forms
@time m = [ uint8(julia(complex(r,i), complex(-.06,.67))) for i=1:-.002:-1, r=-1.5:.002:1.5 ];
# faster than the time from David Sander's demo (newer version of Julia + possibly faster HW)
# elapsed time: 0.240373777 seconds (1502848 bytes allocated)
```

## Type Stabilitly.   When does dynamic behavior begin to fail?

## Forms of this example can be found in most Julia presentations.  The original example is  attributed to Leah Hanson.

### <p>Julia relies on type invariance for performance.</p> <p>Below, we initially assign an int to the variable tot</p> <p>We would not have had a performace problem if  the iterated instruction were tot +=[some integer value]</p> <p>Unfortunately we have an int being added to a float where we have to address type converion. </p> <p>This requires additional memory allocation and garbage collection during each iteration.</p> <p>The compiler currently does not warn for this.</p>

```julia
function ∑(N)        
    tot = 0
    for i in 1:N
        tot += i/2
    end
    return tot  # I do not like the implicit return but will show it in the next funtion
end


function ∑1(N::Int)
    tot = 0.0
    for i in 1:N
        tot += i/2
    end
    tot  # I really don't like this
end

#what about a list comprehension?
function ∑2(N::Int)
    sum ([i/2 for i in 1:N])
end

∑3(N) = sum ([i/2.0 for i in 1:N])  #this seems like a popular idiom



∑4(N::Int) = sum ([i/2.0 for i in 1:N])  #this seems like a popular idiom with a defined type
```

```julia
∑(10), ∑1(10), ∑2(10), ∑3(10), ∑4(10)
```

```julia
methods(∑), methods(∑1), methods(∑2), methods(∑3)
```

```julia
N=1000000
```

```julia
for i in 1:10
    @time(∑(N))
    @time(∑1(N))
    @time(∑2(N))
    @time(∑3(N))
    @time(∑4(N))
    print ("-----------------------------\n")
end
```

```julia
code_typed(∑1,(Int,))
```
